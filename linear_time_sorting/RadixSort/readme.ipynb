### Radix Sort Algorithm

Radix Sort is a non-comparative integer sorting algorithm. It sorts numbers by processing individual digits. The algorithm processes each digit of the numbers, starting from the least significant digit (LSD) to the most significant digit (MSD), or vice versa. Radix Sort is efficient for sorting large lists of numbers and can be implemented using counting sort or bucket sort as subroutines.

#### Steps of Radix Sort:

1. **Determine the Maximum Number of Digits**: Find the number with the maximum digits in the list. This determines the number of passes the algorithm will make.
2. **Sort by Each Digit**: Starting from the least significant digit (LSD), sort the numbers based on that digit. Use a stable sorting algorithm like counting sort or bucket sort to maintain the relative order of numbers with the same digit.
3. **Repeat for Each Digit**: Move to the next significant digit and repeat the sorting process. Continue this until all digits have been processed.

#### Example:

Consider the array `[170, 45, 75, 90, 802, 24, 2, 66]`.

1. **Initial List**: `[170, 45, 75, 90, 802, 24, 2, 66]`
2. **Sort by 1st Digit (units place)**:
    - Buckets: `[170, 90, 802]`, `[2]`, `[24]`, `[45]`, `[75]`, `[66]`
    - Sorted: `[802, 2, 24, 45, 66, 170, 75, 90]`
3. **Sort by 2nd Digit (tens place)**:
    - Buckets: `[802, 2]`, `[24, 45]`, `[66, 75]`, `[170, 90]`
    - Sorted: `[802, 2, 24, 45, 66, 75, 170, 90]`
4. **Sort by 3rd Digit (hundreds place)**:
    - Buckets: `[2, 24, 45, 66, 75, 90]`, `[170]`, `[802]`
    - Sorted: `[2, 24, 45, 66, 75, 90, 170, 802]`

#### Time Complexity:

- **Best Case**: O(d * (n + k)) where d is the number of digits, n is the number of elements, and k is the range of the digit values (typically 0-9 for decimal numbers).
- **Average Case**: O(d * (n + k))
- **Worst Case**: O(d * (n + k))

#### Space Complexity:

- **O(n + k)**: Requires additional space for the output array and the count array.

Radix Sort is an efficient sorting algorithm for numbers and can be faster than comparison-based algorithms like quicksort and mergesort for large datasets. It is a stable sorting algorithm, meaning that it preserves the relative order of equal elements.
