### ExtractMax Operation in Priority Queue

A priority queue is a data structure where each element has a priority, and elements are served based on their priority. The extractMax operation is used to remove and return the element with the highest priority from the priority queue. Here is a step-by-step description of the extractMax algorithm:

1. **Check for Underflow**: Before removing an element, check if the priority queue is empty. If the queue is empty, an underflow condition occurs, and the extractMax operation cannot be performed.
2. **Retrieve the Max Element**: The element with the highest priority is typically at the root of the heap. Retrieve this element.
3. **Replace Root with Last Element**: Replace the root element with the last element in the queue.
4. **Move-Down Function**: Compare the new root with its children. If the new root has a lower priority than one of its children, swap it with the child that has the higher priority. Repeat this process until the new root is in the correct position or it becomes a leaf.
5. **Update the Queue**: The priority queue now no longer contains the removed element, and the heap property is restored.

The extractMax operation typically has a time complexity of O(log n) since it involves moving down the tree, which has a height of log n for a balanced binary heap.
