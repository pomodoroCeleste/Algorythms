### Push Operation in Stack

A stack is a linear data structure that follows the Last In First Out (LIFO) principle. The push operation is used to add an element to the top of the stack. Here is a step-by-step description of the push algorithm:

1. **Check for Overflow**: Before adding an element, check if the stack is full. If the stack is full, an overflow condition occurs, and the push operation cannot be performed.
2. **Increment the Top Pointer**: If the stack is not full, increment the top pointer to point to the next empty space.
3. **Add the Element**: Place the new element in the position pointed to by the top pointer.
4. **Update the Stack**: The stack now contains the new element at the top.

The push operation typically has a time complexity of O(1) since it involves a constant number of steps regardless of the number of elements in the stack.
