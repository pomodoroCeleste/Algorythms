### Pop Operation in Stack

A stack is a linear data structure that follows the Last In First Out (LIFO) principle. The pop operation is used to remove the top element from the stack. Here is a step-by-step description of the pop algorithm:

1. **Check for Underflow**: Before removing an element, check if the stack is empty. If the stack is empty, an underflow condition occurs, and the pop operation cannot be performed.
2. **Retrieve the Top Element**: If the stack is not empty, retrieve the element at the top of the stack.
3. **Decrement the Top Pointer**: Decrement the top pointer to point to the next element below the current top.
4. **Update the Stack**: The stack now no longer contains the removed element at the top.

The pop operation typically has a time complexity of O(1) since it involves a constant number of steps regardless of the number of elements in the stack.
