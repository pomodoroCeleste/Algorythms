### Enqueue Operation in Queue

A queue is a linear data structure that follows the First In First Out (FIFO) principle. The enqueue operation is used to add an element to the rear of the queue. Here is a step-by-step description of the enqueue algorithm:

1. **Check for Overflow**: Before adding an element, check if the queue is full. If the queue is full, an overflow condition occurs, and the enqueue operation cannot be performed.
2. **Increment the Rear Pointer**: If the queue is not full, increment the rear pointer to point to the next empty space.
3. **Add the Element**: Place the new element in the position pointed to by the rear pointer.
4. **Update the Queue**: The queue now contains the new element at the rear.

The enqueue operation typically has a time complexity of O(1) since it involves a constant number of steps regardless of the number of elements in the queue.
