### Initialization of Queue

A queue is a linear data structure that follows the First In First Out (FIFO) principle. The initialization operation is used to set up an empty queue. Here is a step-by-step description of the initialization algorithm:

1. **Create an Empty Array**: Allocate memory for an array that will hold the elements of the queue.
2. **Set Front and Rear Pointers**: Initialize the front and rear pointers to indicate that the queue is empty. Typically, both pointers are set to -1 or 0.
3. **Set Queue Size**: Define the maximum size of the queue to prevent overflow conditions.

The initialization operation typically has a time complexity of O(1) since it involves a constant number of steps regardless of the number of elements in the queue.
