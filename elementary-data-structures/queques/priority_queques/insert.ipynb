## Insertion Algorithm in a Binary Heap with Move Up Function

The insert algorithm in a binary heap involves adding a new element to the heap while maintaining the heap property. Here are the steps:

1. **Add the element to the end of the heap**: Place the new element at the end of the heap (i.e., the last position in the array representation of the heap).

2. **Heapify Up (Move Up)**: Compare the added element with its parent; if the added element is greater (in a max-heap) or smaller (in a min-heap) than its parent, swap them. Repeat this process until the heap property is restored or the element reaches the root.

This process ensures that the heap remains a complete binary tree and maintains the heap property after the insertion of the new element.
