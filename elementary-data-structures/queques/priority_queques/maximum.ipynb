## Finding the Maximum Element in a Max-Heap

The maximum algorithm on a heap is used to find the maximum element in a heap data structure. In a max-heap, the maximum element is always at the root node. The algorithm can be described as follows:

1. **Initialization**: Start at the root node of the heap.
2. **Access the Root**: The root node contains the maximum element in a max-heap.
3. **Return the Maximum**: Return the value of the root node as the maximum element.

In a max-heap, the time complexity to find the maximum element is O(1) because it is always located at the root node.
