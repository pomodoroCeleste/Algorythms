### Stack Vuoto (isEmpty) Operation in Stack

A stack is a linear data structure that follows the Last In First Out (LIFO) principle. The isEmpty operation is used to check if the stack is empty. Here is a step-by-step description of the isEmpty algorithm:

1. **Check the Top Pointer**: Examine the top pointer of the stack.
2. **Determine if Empty**: If the top pointer is at its initial position (typically -1 or 0, depending on the implementation), the stack is empty.
3. **Return Result**: Return a boolean value indicating whether the stack is empty.

The isEmpty operation typically has a time complexity of O(1) since it involves a single comparison regardless of the number of elements in the stack.
