### Insertion Sort Algorithm

Insertion Sort is a simple comparison-based sorting algorithm that builds the final sorted array one item at a time. It is much less efficient on large lists than more advanced algorithms such as quicksort, heapsort, or merge sort. Here is a step-by-step explanation of how the Insertion Sort algorithm works:

#### Steps of Insertion Sort:

1. **Iterate**:
   - Iterate from the second element to the last element of the array.

2. **Compare and Insert**:
   - For each element, compare it with the elements in the sorted portion of the array.
   - Shift all elements in the sorted portion that are greater than the current element to the right.
   - Insert the current element into its correct position in the sorted portion.

#### Example:

Consider the array `[12, 11, 13, 5, 6]`.

1. **First Iteration**:
   - Compare `11` with `12`, shift `12` to the right, and insert `11` at index 0 to get `[11, 12, 13, 5, 6]`.

2. **Second Iteration**:
   - Compare `13` with `12`, no shift needed, `13` stays at index 2.

3. **Third Iteration**:
   - Compare `5` with `13`, shift `13` to the right.
   - Compare `5` with `12`, shift `12` to the right.
   - Compare `5` with `11`, shift `11` to the right.
   - Insert `5` at index 0 to get `[5, 11, 12, 13, 6]`.

4. **Fourth Iteration**:
   - Compare `6` with `13`, shift `13` to the right.
   - Compare `6` with `12`, shift `12` to the right.
   - Compare `6` with `11`, shift `11` to the right.
   - Insert `6` at index 1 to get `[5, 6, 11, 12, 13]`.


#### Time Complexity:

- **Best Case**: O(n) (occurs when the array is already sorted)
- **Average Case**: O(n^2)
- **Worst Case**: O(n^2)

#### Space Complexity:

- **O(1)**: Requires a constant amount of additional space.

Insertion Sort is a simple and intuitive sorting algorithm, but it is not efficient for large lists. It is mainly used for small datasets or as a part of more complex algorithms. It is a stable sorting algorithm, meaning that it preserves the relative order of equal elements.
