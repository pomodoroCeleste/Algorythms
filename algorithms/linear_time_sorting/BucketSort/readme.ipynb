### Bucket Sort Algorithm

Bucket Sort is a comparison sort algorithm that operates on elements by dividing them into different buckets and then sorting these buckets individually. It is particularly useful for sorting a large number of elements that are uniformly distributed over a range.

#### Steps of Bucket Sort:

1. **Create Buckets**: Create an array of empty buckets.
2. **Distribute Elements**: Iterate through the input array and distribute the elements into the appropriate buckets based on a hashing or mapping function.
3. **Sort Buckets**: Sort each non-empty bucket individually. This can be done using any sorting algorithm, such as Insertion Sort or Quick Sort.
4. **Concatenate Buckets**: Concatenate the sorted buckets to form the final sorted array.

#### Example:

Consider an array of floating-point numbers between 0 and 1: `[0.78, 0.17, 0.39, 0.26, 0.72, 0.94, 0.21, 0.12, 0.23, 0.68]`.

1. **Create Buckets**: Create 10 empty buckets (for simplicity, one for each range of 0.1).
2. **Distribute Elements**:
    - 0.78 goes to bucket 7
    - 0.17 goes to bucket 1
    - 0.39 goes to bucket 3
    - 0.26 goes to bucket 2
    - 0.72 goes to bucket 7
    - 0.94 goes to bucket 9
    - 0.21 goes to bucket 2
    - 0.12 goes to bucket 1
    - 0.23 goes to bucket 2
    - 0.68 goes to bucket 6
3. **Sort Buckets**: Sort each bucket individually.
4. **Concatenate Buckets**: Combine the sorted buckets to get the final sorted array: `[0.12, 0.17, 0.21, 0.23, 0.26, 0.39, 0.68, 0.72, 0.78, 0.94]`.

#### Time Complexity:

- **Best Case**: O(n + k) where n is the number of elements and k is the number of buckets.
- **Average Case**: O(n + k + n log(n/k)) assuming the elements are uniformly distributed.
- **Worst Case**: O(n^2) if all elements are placed in a single bucket.

Bucket Sort is efficient for sorting elements that are uniformly distributed over a range and can be parallelized easily.
