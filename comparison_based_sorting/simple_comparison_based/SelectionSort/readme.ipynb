### Selection Sort Algorithm

Selection Sort is a simple comparison-based sorting algorithm that divides the input array into two parts: the subarray of items already sorted, which is built up from left to right at the front (left) of the array, and the subarray of items remaining to be sorted that occupy the rest of the array. Here is a step-by-step explanation of how the Selection Sort algorithm works:

#### Steps of Selection Sort:

1. **Iterate**:
   - Iterate from the first element to the last element of the array.

2. **Find the Minimum**:
   - For each position in the array, find the minimum element in the unsorted portion of the array.

3. **Swap**:
   - Swap the found minimum element with the first element of the unsorted portion of the array.

#### Example:

Consider the array `[64, 25, 12, 22, 11]`.

1. **First Iteration**:
   - Find the minimum element in `[64, 25, 12, 22, 11]`, which is `11`.
   - Swap `11` with `64` to get `[11, 25, 12, 22, 64]`.

2. **Second Iteration**:
   - Find the minimum element in `[25, 12, 22, 64]`, which is `12`.
   - Swap `12` with `25` to get `[11, 12, 25, 22, 64]`.

3. **Third Iteration**:
   - Find the minimum element in `[25, 22, 64]`, which is `22`.
   - Swap `22` with `25` to get `[11, 12, 22, 25, 64]`.

4. **Fourth Iteration**:
   - Find the minimum element in `[25, 64]`, which is `25`.
   - Swap `25` with `25` to get `[11, 12, 22, 25, 64]`.

5. **Fifth Iteration**:
   - The last element `64` is already in its correct position.


#### Time Complexity:

- **Best Case**: O(n^2)
- **Average Case**: O(n^2)
- **Worst Case**: O(n^2)

#### Space Complexity:

- **O(1)**: Requires a constant amount of additional space.

Selection Sort is a simple and intuitive sorting algorithm, but it is not efficient for large lists. It is mainly used for small datasets or as a part of more complex algorithms. It is not a stable sorting algorithm, meaning that it does not preserve the relative order of equal elements.
