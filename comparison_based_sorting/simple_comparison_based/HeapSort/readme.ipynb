### Heap Sort Algorithm

Heap Sort is a comparison-based sorting algorithm that uses a binary heap data structure to sort elements. It works by building a max heap from the input data, then repeatedly extracting the maximum element from the heap and rebuilding the heap until all elements are sorted. Here is a step-by-step explanation of how the Heap Sort algorithm works:

#### Steps of Heap Sort:

1. **Build a Max Heap**:
   - Convert the input array into a max heap, where the largest element is at the root.

2. **Extract Elements**:
   - Repeatedly extract the maximum element from the heap and move it to the end of the array.
   - Reduce the size of the heap and rebuild the heap to maintain the max heap property.

#### Example:

Consider the array `[4, 10, 3, 5, 1]`.

1. **Build a Max Heap**:
   - Convert the array into a max heap: `[10, 5, 3, 4, 1]`.

2. **Extract Elements**:
   - Swap `10` with `1` to get `[1, 5, 3, 4, 10]`.
   - Rebuild the heap to get `[5, 4, 3, 1, 10]`.
   - Swap `5` with `1` to get `[1, 4, 3, 5, 10]`.
   - Rebuild the heap to get `[4, 1, 3, 5, 10]`.
   - Swap `4` with `1` to get `[1, 3, 4, 5, 10]`.
   - Rebuild the heap to get `[3, 1, 4, 5, 10]`.
   - Swap `3` with `1` to get `[1, 3, 4, 5, 10]`.
   - Rebuild the heap to get `[1, 3, 4, 5, 10]`.

#### Time Complexity:

- **Best Case**: O(n log n)
- **Average Case**: O(n log n)
- **Worst Case**: O(n log n)

#### Space Complexity:

- **O(1)**: Requires a constant amount of additional space.

Heap Sort is an efficient sorting algorithm with a time complexity of O(n log n) in all cases. It is not a stable sorting algorithm, meaning that it does not preserve the relative order of equal elements. However, it is widely used due to its efficiency and simplicity.

### Build Heap Algorithm

The `buildHeap` algorithm is a key component of the Heap Sort algorithm. It converts an unsorted array into a max heap, where the largest element is at the root. This is done by calling the `heapify` function on each non-leaf node, starting from the bottom. Here is a step-by-step explanation of how the `buildHeap` algorithm works:

#### Steps of Build Heap:

1. **Start from the Bottom**:
   - Begin with the last non-leaf node and move upwards to the root.

2. **Heapify Each Node**:
   - For each node, call the `heapify` function to ensure the subtree rooted at that node satisfies the max heap property.

#### Example:

Consider the array `[4, 10, 3, 5, 1]`.

1. **Initial List**: `[4, 10, 3, 5, 1]`
2. **Start from the Bottom**:
   - The last non-leaf node is at index `1` (value `10`).

3. **Heapify Each Node**:
   - Call `heapify` on index `1`:
     - The subtree rooted at `10` is already a max heap.
   - Call `heapify` on index `0`:
     - Swap `4` with `10` to get `[10, 4, 3, 5, 1]`.
     - Call `heapify` on index `1`:
       - Swap `4` with `5` to get `[10, 5, 3, 4, 1]`.

The array is now a max heap: `[10, 5, 3, 4, 1]`.

#### Time Complexity:

- **Best Case**: O(n)
- **Average Case**: O(n)
- **Worst Case**: O(n)

#### Space Complexity:

- **O(1)**: Requires a constant amount of additional space.

The `buildHeap` algorithm is efficient and works well with the Heap Sort algorithm. It ensures that the array satisfies the max heap property, which is essential for the subsequent steps of Heap Sort.

### Heapify Algorithm

The `heapify` algorithm is a key component of the Heap Sort algorithm. It ensures that a subtree rooted at a given node satisfies the max heap property. If the subtree does not satisfy the max heap property, the algorithm swaps elements to correct the order and recursively applies the same process to the affected subtree. Here is a step-by-step explanation of how the `heapify` algorithm works:

#### Steps of Heapify:

1. **Identify the Largest Element**:
   - Compare the root element with its left and right children.
   - Identify the largest element among the root and its children.

2. **Swap and Recursively Heapify**:
   - If the root is not the largest, swap it with the largest element.
   - Recursively apply the `heapify` function to the affected subtree.

#### Example:

Consider the array `[4, 10, 3, 5, 1]` and the subtree rooted at index `0` (value `4`).

1. **Initial Subtree**: `[4, 10, 3, 5, 1]`
2. **Identify the Largest Element**:
   - Compare `4` with its left child `10` and right child `3`.
   - The largest element is `10`.

3. **Swap and Recursively Heapify**:
   - Swap `4` with `10` to get `[10, 4, 3, 5, 1]`.
   - Recursively apply `heapify` to the subtree rooted at index `1` (value `4`):
     - Compare `4` with its left child `5` and right child `1`.
     - The largest element is `5`.
     - Swap `4` with `5` to get `[10, 5, 3, 4, 1]`.

The subtree now satisfies the max heap property.

#### Time Complexity:

- **Best Case**: O(log n)
- **Average Case**: O(log n)
- **Worst Case**: O(log n)

#### Space Complexity:

- **O(1)**: Requires a constant amount of additional space.

The `heapify` algorithm is efficient and works well with the Heap Sort algorithm. It ensures that the subtree rooted at a given node satisfies the max heap property, which is essential for the subsequent steps of Heap Sort.
